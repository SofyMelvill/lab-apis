```python
# Your answer here

# functio to get the top artists

def get_top_tracks(artists):
    top_tracks_list = []
    for artist in artists:
        result = sp.search(q=f'artist:{artist}', type='artist', limit=1)
        if not result["artists"]["items"]:
            print(f"Artist '{artist}' not found.")
            continue

        artist_id = result["artists"]["items"][0]["id"]
        top_tracks = sp.artist_top_tracks(artist_id)

        top_tracks_list.extend(top_tracks['tracks'][:5])
    return top_tracks_list
    
artists = ["Ludovico Einaudi", "Bon Iver", "Silva"]
top_5_tracks = get_top_tracks(artists)

for track in top_5_tracks:
    print(f"{track['name']} - {track['artists'][0]['name']}")

    
    
```

```python
def get_related_artists_real(artist_name, limit=5):
    result = sp.search(q=f'artist:{artist_name}', type='artist', limit=1)
    items = result.get("artists", {}).get("items", [])
    if not items:
        print(f"Artist '{artist_name}' not found.")
        return []

    artist_id = items[0]["id"]
    related = sp.artist_related_artists(artist_id)
    return [a['name'] for a in related['artists'][:limit]]
    

print(get_related_artists_real(["Silva"]))
```

```python
def find_related_artists(artists_names, related_limit=3):
    related_artists = []

    for artist in artists_names: 
        try:
            result = sp.search(q=f'artist:{artist}', type='artist', limit=1)
            items = result.get("artists", {}).get("items", [])
            if not items:
                print(f"Artista '{artist}' não encontrado.")
                continue

            artist_id = items[0]["id"]
            related = sp.artist_related_artists(artist_id)
            related_names = [a["name"] for a in related["artists"][:related_limit]]
            related_artists.extend(related_names)

        except spotipy.exceptions.SpotifyException as e:
            print(f"Erro ao buscar relacionados para '{artist}': {e}")
            continue

    seen = set()
    return [a for a in related_artists if not (a in seen or seen.add(a))]

# Test
print(find_related_artists(["Bon Iver"]))
```

```python
def search_playlists(query, limit=5):
    try:
        results = sp.search(q=query, type='playlist', limit=limit)
        playlists = results.get('playlists', {}).get('items', [])

        if not playlists:
            print(f"Nenhuma playlist encontrada para '{query}'.")
            return

        for playlist in playlists:
            name = playlist.get('name', 'Sem nome')
            owner = playlist.get('owner', {}).get('display_name', 'Desconhecido')
            print(f"{name} by {owner}")

    except Exception as e:
        print(f"Erro ao procurar playlists: {e}")

search_playlists("chill")
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```

```python
def get_featured_playlists(limit=5):
    featured = sp.featured_playlists(limit=limit)
    playlists = featured["playlists"]["items"]
    top_playlists = [(pl["name"], pl["id"]) for pl in playlists]
    return top_playlists

top_playlists = get_featured_playlists()
print("\n🎧 Top 5 Featured Playlists:")
for name, pid in top_playlists:
    print(f"- {name} (ID: {pid})")
```

```python
chosen_playlist_id = top_playlists[0][1]  

def get_playlist_details(playlist_id):
    details = sp.playlist(playlist_id)
    name = details["name"]
    description = details["description"]
    total_tracks = details["tracks"]["total"]
    return name, description, total_tracks

name, desc, total = get_playlist_details(chosen_playlist_id)
print(f"\n🔍 Deep Dive:\nNome: {name}\nDescrição: {desc}\nTotal de músicas: {total}")
```

```python
def get_first_10_tracks(playlist_id):
    results = sp.playlist_tracks(playlist_id, limit=10)
    return [track["track"]["name"] for track in results["items"]]

first_10_track_names = get_first_10_tracks(chosen_playlist_id)
print("\n🎵 Primeiras 10 músicas:")
for t in first_10_track_names:
    print(f"- {t}")
```

```python
def get_track_artist_dict(playlist_id, limit=10):
    results = sp.playlist_tracks(playlist_id, limit=limit)
    track_artist_dict = {}
    for item in results["items"]:
        track = item["track"]
        name = track["name"]
        artists = [artist["name"] for artist in track["artists"]]
        track_artist_dict[name] = artists
    return track_artist_dict

track_dict = get_track_artist_dict(chosen_playlist_id)
print("\n🎨 Track e Artistas:")
for track, artists in track_dict.items():
    print(f"- {track}: {artists}")
```
